```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np
```

```python
filename ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
#create headers for the dataframe
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
df.head()
```

```python
#code for replacing one thing with another
df.replace("?",np.nan, inplace = True)
df
```

```python
#evaluating for missing data
missing_data = df.isnull()
missing_data.head(3)
```

```python
#using a For loop to count missing data
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")  
#this shows there are missing values in normalised losses, number of doors, bore, stroke, horsepower, peak rpm and price.
```

<b><u>DEALING WITH MISSING VALUES</u></b><br>
Depending on the context, caution should be exercised calculating means, and the impact on results should be considered.<br>
Whole columns should only be dropped if they are mostly empty.<br>
For the purpose of this wrangling project to showcase coding ability, the missing data will be replaced with means, one by the most freqent, and one column will be dropped. 

```python
avgnormloss = df["normalized-losses"].astype("float").mean(axis = 0)
print("Average of normalized-losses:", avgnormloss)

#replace NaN with avgnormloss in the normalized-losses column
df["normalized-losses"].replace(np.nan, avgnormloss, inplace = True)
df.head(2)
```

```python
avgbore = df["bore"].astype("float").mean(axis = 0)
print("Average bore length is:", avgbore)

df["bore"].replace(np.nan, avgbore, inplace = True)
```

```python
avghorsepower = df["horsepower"].astype("float").mean(axis=0)
print("Average Horsepower is:", avghorsepower)

df["horsepower"].replace(np.nan, avghorsepower, inplace = True)
```

```python
avgstroke = df["stroke"].astype("float").mean(axis = 0)
print("Average stroke length is:", avgstroke)

df["stroke"].replace(np.nan, avgstroke, inplace = True)
```

```python
avgpeakrpm = df["peak-rpm"].astype("float").mean(axis = 0)
print("Average peak RPM length is:", avgpeakrpm)

df["peak-rpm"].replace(np.nan, avgpeakrpm, inplace = True)
```

```python
#replace missing average doors with most frequent
df['num-of-doors'].value_counts()
#most common value is "four"
df["num-of-doors"].replace(np.nan, "four", inplace = True)
```

```python
#drop price
df.dropna(subset=["price"], axis=0, inplace = True)

# reset index, because a row was dropped, and this will change indexing. 
df.reset_index(drop=True, inplace=True)
df.head()
```

We could run the For loop again, to confirm there are no True values remaining in the dataset.<br>
We can see below there are now only False values in the dataset, meaning there is no missing data.

```python
missing_data_update = df.isnull()
missing_data_update.head(3)
for column in missing_data_update.columns.values.tolist():
    print (missing_data_update[column].value_counts())
    print("")  
```

<b><u>CORRECTING DATA FORMAT</b></u><br>
The last step in data cleaning is checking and making sure that all data is in the correct format, so that analysis can be performed.

```python
df.dtypes
```

```python
#change data types
df[["bore", "stroke", "peak-rpm", "price"]] = df[["bore", "stroke", "peak-rpm", "price"]].astype("float")
df[["normalized-losses", "horsepower"]] = df[["normalized-losses", "horsepower"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df.dtypes
```

<b><u>DATA STANDARDISATION & NORMALISATION</B></U><br>
Allows for easier, and more meaningful clear comparisons of data.<br>
Here, data will be normalised so their value ranges from 0 to 1.

```python
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df["height"] = df["height"]/df["height"].max()
df[["length","width","height"]].head()
```

<b>Binning</b><br>
Horsepower is a continous feature of the dataset.<br>
We can separate the values into categories. For this example, values can be binned into low, medium, high horsepower.<br>
The feature horsepower has already been coverted into the correct data format (int) as above.

```python
horsepowerbins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bin_names = ['Low', 'Medium', 'High']

#Use "cut" function which to apply each value to a bin.
df["horsepower-binned"] = pd.cut(df["horsepower"], horsepowerbins, labels = bin_names, include_lowest = True)

#see how many vehicles in each bin
df["horsepower-binned"].value_counts()
```

```python
#A bar chart of the number of cars in each bin
plt.bar(bin_names, df["horsepower-binned"].value_counts(), color = "#FBD048", edgecolor = "#000000")

plt.xlabel("Horsepower")
plt.ylabel("Number of cars in bin")
plt.title("Horsepower Bins Quantity")
plt.show()
```

<b>HISTOGRAM</b>

```python
plt.hist(df["horsepower"], bins = 3, color = "#4891FB", edgecolor = "#000000")

plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower Bins")
plt.show()
```

<b> INDICATOR VARIABLES </b> <br>
Numerical variables used to label categories, the numbers do not have inherent meaning.<br>
They are used so that you can use categorical variables for regression analysis in the later modules.<br>
E.g: fuel type has two variables: gas and diesel. These could be used for regression analysis, but for this type of analysis,<br>
the values must be numberical.

```python
fuelindicator = pd.get_dummies(df["fuel-type"])
fuelindicator.head(3)               
```

```python
#rename columns for clarity.
fuelindicator.rename(columns={"diesel":"fuel-type-diesel","gas":"fuel-type-gas"}, inplace = True)
fuelindicator.head(3)
```

```python
#merge df and the indicator
df = pd.concat([df, fuelindicator], axis = 1)

df.head()
```

```python
#drop original fuel type column
df.drop("fuel-type", axis = 1, inplace = True)
df.head()
```
